````python
# TASK: Guessing a number

answer = 27
guess = int(input("Guess the number: "))

while guess != answer:
    while guess < answer:
        print("It is more than your guess. Try again!")
        guess = int(input("Guess the number: "))
    
    while guess > answer:
        print("It is less than your guess. Try again!")
        guess = int(input("Guess the number: "))
    
print("Congrats! you got it right. The number was 27.")
````

````python
# TASK: Guessing a number (TEACHER'S SOLUTION)

answer = 27

print("Let's play guessing the number")

while True:
    guess = int(input("Guess the number: "))
    if guess < answer:
        print("It is more than your guess. Try again!")
    elif guess > answer:
        print("It is less than your guess. Try again!")
    else:
        print("Congrats! you got it right. The number was 27.")
        break
````

````python
# TASK: Longest Word

sentence = input("Write a sentence to find its longest word: ")

word_list = sentence.split()

longest = 0
i = 0

while i < len(word_list):
    if len(word_list[i]) > longest:
        longest = len(word_list[0])

    i += 1
print("The length of the longest word: ", longest)
````

````python
sentence = input("Write a sentence to find its longest word: ")

words = sentence.split()
longest = 0
i = 0

while i < len(words):
    if len(words[i]) > longest:
        longest = len(words[i])
        
    i += 1


print(f"The length of the longest word is {longest}")
````

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Nested For Loop</p>

````python
#TASK (Created indicidually)

# Write a program to create all possible combinations of x : y for x in numbers and y in fruit
# i.e 1:apple, 2:apple, 3:apple, 4:apple, 1:banana, 2:banana, ...., 


fruit = ['apple', 'banana', 'orange']
numbers = [1, 2, 3, 4]

for n in fruit:
    for i in numbers:
        print(i, ":", n)
````

````python
# TO CHECK IF A NUMBER IS PRIME

n = int(input('Enter a number to check if it is a prime number : '))

no_of_factors = 0

for i in range(1,n+1):
    if n % i == 0:
        no_of_factors += 1

if no_of_factors == 2:
    print(f"{n} is a prime number")
else:
    print(f"{n} is NOT a prime number")
````

````python
# LISTING PRIME NUMBERS FROM 1 TO 500 (My Solution)

primes = []  # the empty list to append the primes

for n in range (2, 101):
    no_of_factors = 0
    for i in range(1,n+1):
        if n % i == 0:  
            no_of_factors += 1
    if no_of_factors == 2: # check if it has exactly two factors
        primes.append(n) # append to primes if it has exactly two primes
        
print(primes)
````

````python
# LISTING PRIME NUMBERS FROM 1 TO 500 (Solution in LMS)

prime=[]  # created an empty list to collect prime numbers in it

for num in range(2, 101):
    status = True
    for i in range(2, num):
        if num % i == 0: # check if the only factors are 1 and itself
            status = False
    if status:
        prime.append(num)  # collect prime numbers in the list

print(prime)
````
